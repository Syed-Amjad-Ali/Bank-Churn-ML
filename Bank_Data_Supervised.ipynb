```python
# Convert all boolean columns to integers (0 or 1)
for col in modelling_df.select_dtypes(include=["bool"]).columns:
    modelling_df[col] = modelling_df[col].astype(int)


modelling_df=pd.get_dummies(modelling_df,drop_first=True, dtype="int")

modelling_df.head()
```

```python
modelling_df.info()
```

```python
## spliiting the dataq
```

```python
from sklearn.model_selection import train_test_split

x = modelling_df.drop("Exited", axis=1)
y = modelling_df["Exited"]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2024)


x_train.info()
```

```python
x_test.info()
```

```python
#objective 3
#build and evaluate logistic regression
#build a confusion matrix
#accuracy, preciosion , recall and f1
#roc curve and calculate AUC statistic
#plot precision and recall against the model threshold
```

```python
# from sklearn.linear_model import LogisticRegression

# logreg = LogisticRegression() #increasing iter helps model find solution

# lr = logreg.fit(x_train,y_train)

# print(f"Train Accuracy:{lr.score(x_train,y_train)}" )




from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(solver= "newton-cholesky") #increasing iter helps model find solution

lr = logreg.fit(x_train,y_train)

print(f"Train Accuracy:{lr.score(x_train,y_train)}" )
```

```python
from sklearn.metrics import confusion_matrix,precision_score, recall_score,f1_score

confusion_matrix(y_train,lr.predict(x_train))
```

```python
precision_score(y_train,lr.predict(x_train))
```

```python
#now we do the roc curve

from sklearn.metrics import roc_curve, auc

y_probs = lr.predict_proba(x_test)[:,1]

fpr1, tpr1, threshholds = roc_curve(y_test,y_probs)

auc_score = auc(fpr1,tpr1)

#plot the ROC curve
plt.plot(fpr1,tpr1, label=f'LR (AUC = {auc_score:.2f})')

#Draw Random Guess
plt.plot([0,1], [0,1], 'k--', label = 'Random Guess (AUC = 0.5)')

#Modify Formatting
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Bank Churn Model')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

p_curve, r_curve, t_curve = precision_recall_curve(y_train,lr.predict_proba(x_train)[:,1])

plt.plot(t_curve, p_curve[:-1],label='Precision')
plt.plot(t_curve,r_curve[:-1], label ='Recall')
plt.xlabel('Prediction Threshold')
plt.ylabel('Scores')
plt.legend()
plt.title('Precision & Recall Curves')
plt.show()
```

```python
#so our goal is to find the pooint where precision stays above 0.5
#roughly around the point of intersection and the goal is to 
#maximize the recall. so we can capture 50% churners with relatively small
#sacrifice in precision
# so we can define point where precision >0.5
```

```python
start_time = time.time()

from sklearn.model_selection import GridSearchCV

params= {
    'n_estimators': np.arange(start=250, stop =351, step =10),
    'max_samples': [0.4 , .5, .6],
    'max_depth': [6,8,10],
    'min_samples_leaf':[5,10,15]
    
}

grid = GridSearchCV(
    rf,
    params,
    scoring="accuracy"
)

grid.fit(x_train,y_train)
grid.best_params_ 

end_time = time.time()

print("Best parameters:")
print(grid.best_params_)


print(f"Total runtime: {(end_time - start_time) / 60:.2f} minutes")
```
